```python
def get_min_max(df):
    min_max_list = []
    for col in df.columns:
        min_val = df[col].min()
        max_val = df[col].max()
        min_max_list.append(f"{col}({min_val},{max_val})")
    return min_max_list
```

Func untuk mencari nilai minimum dan maximum tiap kolom

```python
get_min_max(df)
```
